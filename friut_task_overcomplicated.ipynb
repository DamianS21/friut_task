## Task:
* People: 22
* Apples: 16
* Bananas: 14
* Pears: 18

### Script

```python
PEOPLE = 22
CONST_FRIUTS = [16,14,18]
N_FRIUTS = len(CONST_FRIUTS)
friuts = [[0]*PEOPLE for x in range(N_FRIUTS)]
div_list = [0]*N_FRIUTS


def assign_all(list0,i):
    for j in range(len(friuts)):
        list0[j][i] = 1


def recurrent(friuts,i, max_i):
    if(i < max_i):
        assign_all(friuts, i)
        for y in range(N_FRIUTS):
            div_list[y] = CONST_FRIUTS[y] - sum(friuts[y])
        friut_sum = sum(div_list)
        if friut_sum == max_i - i:
            return i
        elif friut_sum <= max_i - i:
            #print('overloaded - no more friuts')
            return i-1
        return recurrent(friuts,i+1, max_i)

result = recurrent(friuts, 0, PEOPLE-1)
print('Maximum people with '+str(N_FRIUTS)+ ' friuts: '+str(result+1))
```
